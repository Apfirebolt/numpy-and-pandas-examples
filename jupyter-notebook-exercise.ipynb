```python
import pandas as pd
import numpy as np

df = pd.read_csv('HDFCBANK.csv')
rows = df.head(10)

# Convert the dataframe into pure dictionary, each column is key
rows = dict(rows)
print(rows['Close'])
```

```python
anotherDf = df.head(10)
for index, item in anotherDf.iterrows():
    print(item['Open'])
```

```python
columns = df.columns
print(columns)
print(df.head(5))
```

```python
anotherDf = df.head(10)
data = []
for index, item in anotherDf.iterrows():
    currentDict = {}
    for column in columns:
        currentDict[column] = item[column]
    data.append(currentDict)

for oneData in data:
    print('Data - ', oneData)
```
